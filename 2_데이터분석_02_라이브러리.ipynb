### 2-3.Numpy
#### #Numpy의 특징

#### 1.행렬 데이터를 표현할 수 있다.


```python
import numpy as np

x1 = np.array([5, 3, 1, 0.2])  # 하나의 행렬에 하나의 데이터 타입만 가질 수 있다
print(x1)
print( type(x1) )
print(f'차원: {x1.ndim}')
print(f'구조: {x1.shape}')

x2 = np.array([[5], [3], [1], [0.2]])
print(x2)
print( type(x2) )
print(f'차원: {x2.ndim}')
print(f'구조: {x2.shape}')
```

```python
import numpy as np

arr = np.zeros((2,2)) # 2행 2열 행렬에 값을 모두 0으로 채운다.
print(arr)

arr = np.ones((2,3)) # 2행 3열 행렬에 값을 모두 1으로 채운다.
print(arr)

arr = np.full((2,3), 255) # 2행 3열 행렬에 값을 모두 255으로 채운다.
print(arr)

arr = np.eye(3)  # 3행 3열의 주 대각행렬 요소의 값을 1로 채운다.
print(arr)

arr=np.array(range(20)).reshape((4,5)) # 0~19를 4행 5열 행렬로 만든다.
print(arr)

s = np.sum(arr) # arr행렬 원소 모두 합산하기
print(s)

s = np.sum(arr, axis=0) # 세로축 합산
print(s)

s = np.sum(arr, axis=1) # 가로축 합산
print(s)
```

#### 1.선(Line) 그래프

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (5, 3)        # 그래프 크기(인치)
plt.rcParams['axes.unicode_minus'] = False     # 그래프 축 마이너스 표시


X = list(range(2005, 2015, 1))  # X좌표
Y = [15,7,2,10,8,5,14,9,18,8]   # Y좌표
print('X축 값: ', X)
print('Y축 값: ', Y)

# plt.plot(X, Y)  # 기본 선 그래프
# plt.plot(X, Y, linestyle='dashdot', color='r') # 빨간색 점선 그래프
plt.plot(X, Y, color='lavender', marker='^', linestyle='solid',
         linewidth=2, markersize=10, markerfacecolor='salmon')
plt.grid(True)
plt.title('연도별 그래프')
plt.xlabel('연도')
plt.ylabel('지진강도')
plt.show()
```

- 2차 방정식의 그래프 : f(x) = a*x^2 + b, 포물선 방정식

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-4.5,5,0.5)    # numpy에서 range 함수가 arange.
y = 2*x**2
print('X축 값: ', x)
print('Y축 값: ', y)

plt.plot(x, y, label='L1')
plt.plot(x, 4*x, label='L2')
plt.plot(x,-3*x, label='L3')
plt.legend()
plt.show()
```

- 파일 데이터 읽어서 선 그래프 그리기

```python
import pandas as pd
import matplotlib.pyplot as plt

파일이름 = '/content/sample_data/weather_total.csv'
df = pd.read_csv(파일이름, encoding='cp949')

#컬럼명 변경하기
df.columns = ['일시','기온','습도','구름','일조시간','강수시간','구분']

# # 1.년도 추출: 일시에서 년도만 자르기
df['년도'] = df['일시'].str[0:4]
df['년도'] = pd.to_datetime(df['일시']).dt.year
df
# 2. query문에서 특정 년도(2010)가 들어간 부분 검색하기
df2 = df.query(" 일시.str.contains('01-01') ")   # 조건식
X = df2['일시']
Y1 = df2['일조시간']
Y2 = df2['강수시간']

plt.plot(X, Y1,  color= 'red',marker='o', label='일조시간')
# plt.plot(X, Y1, '*-', color='red', label='일조시간')
plt.bar(X, Y2, label='강수시간')

plt.title('연도별 일조시간&강수시간', fontsize=15)
plt.xlabel('연도')
plt.ylabel('시간')
plt.xticks(rotation=45)
plt.legend()

plt.show()
```

#### 2.점(scatter) 그래프
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

```python
import matplotlib.pyplot as plt

x = list(range(2005,2015,1))
y  = [15,7,2,10,8,5,14,9,18,8]

plt.scatter(x, y)
plt.show()
```

- 점 크기 지정

```python
plt.scatter(x, y,  s=500, c='red', alpha=0.4)  # s: 마커크기, alpha: 투명도
for a, b in zip(x, y):
    plt.text(a, b, str(b), fontsize=9,
             horizontalalignment='center', verticalalignment='center')  # x,y좌표에 글자 넣기
plt.show()
```

- 점 크기, 색상, 투명도 지정

```python
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(12345678)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

#### 3.막대(bar) 그래프

```python
import matplotlib.pyplot as plt

# 그래프 크기 설정
plt.rcParams['figure.figsize'] = (6, 3)  # (가로,세로) 인치 단위

# X축, Y축 데이터
X = list(range(2005,2015,1))
Y = [15,7,2,10,8,5,14,9,18,8]

plt.bar(X, Y)     # 막대 그래프
plt.grid(False)
plt.title('연도별 그래프')
plt.show()
```

- 가로막대

```python
plt.barh(X, Y)     # 가로 막대 그래프
plt.grid(True)
plt.title('연도별 그래프')
plt.show()
```

- 그래프 색상 지정

```python
#그래프 색상 지정
colors = ['r','g','b','c','m','y','k','w']  # 기본색상  # Hexa코드 or CSS컬러사용
plt.bar(X, Y,color=colors)
plt.show()
```

```python
# plt.bar(X, Y) # 막대 그래프
# plt.barh(X, Y) # 막대(가로) 그래프
Y2 = [i-2 for i in Y]
plt.bar(X, Y, label='값1')    # 파란색
plt.bar(X, Y2, label='값2')    # 주황색 (뒤에 쓰는게 앞에 표시됨.)

plt.grid(False)   # 그리드 출력 여부
plt.title('연도별 그래프')
plt.xlabel('연도')
plt.ylabel('값')
plt.legend()
plt.show()
```

-  막대 2개 나란히 나타내기

```python
import matplotlib.pyplot as plt
import numpy as np

# X축, Y축 데이터
X = list(range(2005,2015,1))
Y1= [15,7,2,10,8,5,14,9,18,8]
Y2= [y+1 for y in Y1]
lable = X
X = np.arange(len(X))

plt.bar(X-0.2, Y1, width=0.4, label='Y1')     # 막대 그래프
plt.bar(X+0.2, Y2, width=0.4, label='Y2')     # 막대 그래프
plt.xticks(X, lable)

plt.title('년도별 그래프')
plt.grid(False)
plt.legend()
plt.show()
```

#### 4.원 그래프

```python
import matplotlib.pyplot as plt
import pandas as pd

idx = ['1학년','2학년','3학년','4학년']
val = [16,12,7,5]
per = [val[x]/sum(val) *100 for x in range(len(val))]

plt.pie(per, autopct='%1.1f%%',textprops=dict(color="w"))
plt.title("학년별 동아리 회원수")
plt.legend(idx, title='학년', loc="center right",
           bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()
```

#### 5.혼합그래프 : 선 + 막대

```python
# X축, Y축 데이터
X = list(range(2005,2015,1))
Y1= [15,7,2,10,8,5,14,9,18,8]
Y2= [y+1 for y in Y1]

plt.plot(X, Y1, 'o', linestyle='dashed', c='r', label='Y1') # 선 그래프
plt.bar(X, Y2, width=0.4, label='Y2')     # 막대 그래프
plt.xticks(X, lable)   # x축 갯수 및 이름

plt.title('연도별 그래프')
plt.grid(False)
plt.legend()
plt.show()
```
