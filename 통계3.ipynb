```python
import pandas as pd
df = pd.read_excel('car.xlsx')
df.head()
```

```python
# 회귀분석
from statsmodels.formula.api import ols

m = ols('price~year', data=df).fit()
m.summary()
```

- year가 1년 증가하면 106만원씩 증가한다.
- 귀무가설: 모집단에서 기울기는 0이다.(연식이 높든 낮든 가격과 무관하다.)
 - p-val < 0.05 귀무가설 기각.

```python
m = ols('price ~ model', data=df).fit()
m.summary()
```

- Intercept	833.4146 (Avante 예상가격)
- model[T.K3]	80.3970 (K3와 Avante의 예상가격 차이)

**price = 833 + 80 * model[T.K3]**

- Avante: `model[T.K3] = 0` 833만원으로 예상
- K3: `model[T.K3] = 1` 913만원으로 예상

- p-value > 0.05이므로 귀무가설: "둘의 차이 = 0"을 기각할 수 없음
- 또는 신뢰구간 -10~+171(0도 포함됨)
- 따라서 모집단에서도 K3가 Avante보다 비싸다고 하기에는 증거가 부족하다

```python
avante = df.price[df.model == 'Avante']
k3 = df.price[df.model =='K3']
pg.ttest(k3, avante, correction=False)
# 독립표본 t 검정에서 보정을 안하면 회귀분석과 결과가 같음
# 무엇을 보정하나? 이분산성(heteroscedasticity): 가격대가 퍼져있는 정도가 다르다
```

- 두 집단의 차이가 나느냐, 안 나느냐는 t 검정이 더 정확(왜? 보정을 하니까~)
